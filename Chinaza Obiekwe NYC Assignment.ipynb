```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('NYC_Property_Data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df[df['TAX CLASS AT PRESENT'].isnull()].head()
```

```python
df[df['BUILDING CLASS AT PRESENT'].isnull()].head()
```

```python
df = df.drop( ['TAX CLASS AT PRESENT','BUILDING CLASS AT PRESENT','NEIGHBORHOOD','BUILDING CLASS CATEGORY', 'ADDRESS','SALE DATE'], axis=1)
```

```python
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
df = df.dropna(axis=0, subset=['SALE PRICE', 'LAND SQUARE FEET', 'GROSS SQUARE FEET'])
```

```python
df = df[df['SALE PRICE']>0]
```

```python
df = pd.get_dummies(data=df, columns=['BOROUGH','TAX CLASS AT TIME OF SALE','BUILDING CLASS AT TIME OF SALE'])
```

```python
df.info()
df.isnull().sum()
```

```python
df.corr()['SALE PRICE'].sort_values(ascending=False)
```

```python
#create X and y
X = df.drop('SALE PRICE',axis=1)
```

```python
y = df['SALE PRICE']
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#use Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
dic = {'Actual value' : y_test, 'Predicted Value' :y_pred}
```

```python
Result = pd.DataFrame(dic)
```
