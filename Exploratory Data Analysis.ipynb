```python
# to import multiple csvs in python
import glob

path = './Sales_Data/' # use your path
all_files = glob.glob(path + "/*.csv") # a function that goes through the file that ends with a csv

li = [] # declare an empty list

for filename in all_files: #iterate through all the files in the variable name=all_files
    df = pd.read_csv(filename, index_col=None, header=0)# make the header zero
    li.append(df)

data = pd.concat(li, axis=0, ignore_index=True)
```
